# 2 - Tratamento dos dados:

```python
# Importar planilha - Data de cadastro de título diário

# Importar planilha - Data de cadastro de título diário
data_de_cadastro = pd.read_excel(r"C:\Planilha\Data de cadastro de título diário 1.xls", converters={'Nr. da obrigação':str}) # Importando a planilha definindo o campo 'Nr. da obrigação' como string para manter a formatação do campo 

# Importar planilha - Filtros adicionais 2 diário
filtros_adicionais = pd.read_excel(r"C:\Planilha\Filtros adicionais 2 diário 1.xls", converters={'Nr. da obrigação':str}) # Importando a planilha definindo o campo 'Nr. da obrigação' como string para manter a formatação do campo

# Acerto de tipos de campos diferentes entre as planilhas antes do merge
data_de_cadastro["Agrupamento"] = data_de_cadastro["Agrupamento"].astype(str)

# Combinando arquivos e tratando possíveis duplicidades
df_merge = data_de_cadastro.merge(filtros_adicionais) # Filtrando os itens repetidos

df_merge = data_de_cadastro.merge(filtros_adicionais, how='outer') # Excluindo linhas repetidas

# Apagando as linhas de resultados totais de cada planilha importada
df_merge = df_merge.dropna(subset=['Empresa']) # Apagar linhas em que a coluna 'Empresa' seja "NaN"

# Apagando linhas com a data do dia atual na coluna "Agrupamento":
# Pegando a data atual como índice
indexNames = df_merge[ df_merge['Agrupamento'] == (hoje) ].index
# Deletando essas linhas do dataFrame
df_merge.drop(indexNames , inplace=True)

# Apagar as linhas com a coluna AGRUPAMENTO com datas do dia atual
df_merge = df_merge.dropna(subset=['Empresa'])

# Convertendo os campos "Dt. venc. programado", "Dt. venc. original" e "Dt. cadastro" em string para retirar a hora do campo, ficando só a data
df_merge["Dt. venc. programado"] = df_merge["Dt. venc. programado"].astype(str)
df_merge["Dt. venc. original"] = df_merge["Dt. venc. original"].astype(str)
df_merge["Dt. cadastro"] = df_merge["Dt. cadastro"].astype(str)

# Salvar os dados novos na planilha padrão
filename = 'base/Planilha_merge.xlsx'
df_merge.to_excel(filename, sheet_name='Sales merged', index=False)
```
