# 2 - Tratamento dos dados:

```python
# Importar planilha - Data de cadastro de título diário

# Importar planilha - Data de cadastro de título diário
data_de_cadastro = pd.read_excel(r"C:\Planilha\Data de cadastro de título diário 1.xls") # O "r" é pra que o python não tente ler caracteres especiais como "\", porque o python tem vários caracteres especiais que tem funçãoes específicas


# Importar planilha - Filtros adicionais 2 diário
filtros_adicionais = pd.read_excel(r"C:\Planilha\Filtros adicionais 2 diário 1.xls") # O "r" é pra que o python não tente ler caracteres especiais como "\", porque o python tem vários caracteres especiais que tem funçãoes específicas

# Acerto de tipos de campos diferentes entre as planilhas antes do merge
data_de_cadastro["Agrupamento"] = data_de_cadastro["Agrupamento"].astype(str)

# Combinando arquivos e tratando possíveis duplicidades
df_merge = data_de_cadastro.merge(filtros_adicionais) # Filtrando os itens repetidos

df_merge = data_de_cadastro.merge(filtros_adicionais, how='outer') # Excluindo linhas repetidas

# Apagando as linhas de resultados totais de cada planilha importada
df_merge = df_merge.dropna(subset=['Empresa']) # Apagar linhas em que a coluna 'Empresa' seja "NaN"

# Salvar os dados novos na planilha padrão
filename = 'base/Planilha_merge.xlsx'
df_merge.to_excel(filename, sheet_name='Sales merged', index=False)
```

```python
# df_merge = df_merge.dropna(subset=['Empresa']) # Apagar linhas em que a coluna 'Empresa' seja "NaN" para apagar totais

# Importar planilha padrão - Contas a Pagar:
contas_a_pagar = pd.read_excel(r"C:\Planilha\Contas a Pagar_2020 - 2021.xlsx") # O "r" é pra que o python não tente ler caracteres especiais como "\", porque o python tem vários caracteres especiais que tem funçãoes específicas

# Importar planilha - Merged
planilha_merged = pd.read_excel('base/Planilha_merge.xlsx') 
planilha_merged["BANCO PAGAMENTO"] = ""
planilha_merged["Unnamed: 15"] = ""


# converter colunas dataframe(string no formato data) em dateTime:
planilha_merged["Nr. da obrigação"] = planilha_merged["Nr. da obrigação"].astype(str)
planilha_merged["Cód. pessoa"] = planilha_merged["Cód. pessoa"].astype(str)
planilha_merged["Dt. venc. original"] = planilha_merged["Dt. venc. original"].astype(str)
planilha_merged["Dt. cadastro"] = planilha_merged["Dt. cadastro"].astype(str)

# Combinando arquivos e tratando possíveis duplicidades
df_merge_new = contas_a_pagar.merge(planilha_merged) # Filtrando os itens repetidos

df_merge_new = contas_a_pagar.merge(planilha_merged, how='outer') # Excluindo linhas repetidas

# Salvar os dados novos na planilha padrão
filename_new = 'base/Planilha_completa.xlsx'
df_merge_new.to_excel(filename_new, sheet_name='Sales merged', index=False)
```
